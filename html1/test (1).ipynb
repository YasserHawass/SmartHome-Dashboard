```python
import plotly.express as px
import pandas as pd

# to run this code, using jupyter on VS Code
from plotly.offline import plot
import plotly.io as pio
pio.renderers.default = "notebook_connected"
```

```python
df = px.data.gapminder()
```

```python
df
```

```python
fig = px.scatter(df, x="gdpPercap", y="lifeExp", size="pop", color="continent",
              hover_name="country", log_x=True, size_max=60, animation_frame="year", title="Gapminder", height=650, trendline="ols", template="plotly_dark", range_y=[20, 90])


fig.show() # This will show the plot in the notebook.
# plot(fig)   # This will show the plot in a browser.
              
```

```python
df2 = px.data.stocks()
df2
```

```python
fig2 = px.line(df2, x="date", y="GOOG", title="Google", markers=True)
fig2.show(renderer="notebook_connected")
```

```python
df3 = px.area(df2, x="date", y=["GOOG", "AAPL", "AMZN"], title="Google, Apple, Amazon")
df3.show(renderer="notebook_connected")
```

```python
df4 = px.bar(df2, x="date", y=["GOOG", "AAPL", "AMZN"], title="Google, Apple, Amazon", barmode='group')
df4.show(renderer="notebook_connected")
```

```python
df3 = px.data.tips()
df3
```

```python
fig5 = px.scatter(df3, x="total_bill", y="tip", color="smoker", size="size",
                    hover_name="smoker", log_x=True, size_max=60, animation_frame="day", title="Tips", height=650, trendline="ols", template="plotly_dark", range_y=[0, 60])
fig5.show(renderer="notebook_connected")
```

```python
fig6 = px.pie(df3, values="tip", names="day", title="Tips")
fig6.show(renderer="notebook_connected")
```

```python
fig7 = px.box(df3, x="day", y="tip", title="Tips")

fig7.show(renderer="notebook_connected")
```

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go

fig = make_subplots(rows=1, cols=2)

fig.add_trace(
    go.Scatter(x=[1, 2, 3], y=[4, 5, 6]),
    row=1, col=1
)

fig.add_trace(
    go.Scatter(x=[20, 30, 40], y=[50, 60, 70]),
    row=1, col=2
)

fig.update_layout(height=600, width=800, title_text="Side By Side Subplots")
fig.show()
```
